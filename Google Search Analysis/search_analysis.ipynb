```python
import pandas as pd
import matplotlib.pyplot as plt
from pytrends.request import TrendReq
```

```python
trends = TrendReq()
```

```python
# Analayzing the Google search trends on the query "Machine Learning"
trends.build_payload(kw_list=["Machine Learning"])
```

```python
data = trends.interest_by_region()
df = data.sort_values(by="Machine Learning", ascending=False).head(10).reset_index().copy()
df
```

```python
df.plot(x='geoName', y="Machine Learning", 
          figsize=(15,12), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Machine Learning'])
df = data.interest_over_time()
df
```

```python
fig, ax = plt.subplots(figsize=(15, 12))
df['Machine Learning'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Machine Learning', 
          fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
