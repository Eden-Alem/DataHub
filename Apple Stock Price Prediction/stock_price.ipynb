```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
data = pd.read_csv('./AAPL.csv')
```

```python
data.shape
```

```python
# style of financial chart describining open, high, low and close for a given x coordinate (most likely time)
figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"], high=data["High"],
                                        low=data["Low"], close=data["Close"])])

figure.update_layout(title = "Apple stock price analysis", xaxis_rangeslider_visible = True)
figure.show()
```

```python
# to predict to stock prices of apple for the next 5 days
from autots import AutoTS
# Auto Time Series - Genetic algorithm approach for time series
# Ensemble - techniques that create multiple models and then combine them to produce improved results
model = AutoTS(forecast_length=5, frequency='infer', ensemble='simple')
model = model.fit(data, date_col = 'Date', value_col = 'Close', id_col = None)
model
```

```python
model.results()
```

```python
prediction = model.predict()
```

```python
forecast = prediction.forecast
forecast
```
